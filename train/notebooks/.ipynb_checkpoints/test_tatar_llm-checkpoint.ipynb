```python
import torch
import transformers

model_id = "Tweeties/tweety-tatar-base-7b-2024-v1"
bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model = transformers.AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    quantization_config=bnb_config,
    device_map='auto',
)

tokenizer = transformers.AutoTokenizer.from_pretrained(
    model_id,
)

def generate_response(user_message, context=""):
    prompt = f"Сорау: {user_message}\nкыска җавап:"
    encodeds = tokenizer(prompt, return_tensors="pt", add_special_tokens=False)
    model_inputs = encodeds

    generated_ids = model.generate(**model_inputs, max_new_tokens=200, do_sample=True)
    decoded = tokenizer.batch_decode(generated_ids)
    response = decoded[0].split("<extra_id_0>")[-1].strip()
    return response

while True:
    user_message = input("Enter your message (or type 'exit' to quit): ")
    if user_message.lower() == "exit":
        break
    response = generate_response(user_message)
    print("Model response:")
    print(response)
```
